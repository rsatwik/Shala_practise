```python
import os
path = '/content/drive/MyDrive/image_processing'

imglist = []
for i in os.scandir(path):
  imglist.append(i.path)

imglist
```

```python
from matplotlib import image
import matplotlib.pyplot as plt
im = image.imread(imglist[0])
plt.imshow(im)
plt.show()
```

```python
import cv2
```

```python
im2 = cv2.blur(im,(15,15)) # blurring the image
plt.imshow(im2)
plt.show()
```

```python
from skimage.util import random_noise

im2 = random_noise(im,mode='s&p',amount=0.2) # adding salt and pepper noise
plt.imshow(im2)
plt.show()
```

```python
# Convert image to grayscale
gim = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(gim)
```

```python
# using cv2 imread to read image
gim = cv2.imread(imglist[0])
plt.imshow(gim)
```

```python
# Convert image to grayscale
gim = cv2.imread(imglist[0],0)
plt.imshow(gim)
```

```python
# calculate frequency of pixels in range 0-255
histg = cv2.calcHist([gim],[0],None,[256],[0,256]) # https://www.geeksforgeeks.org/opencv-python-program-analyze-image-using-histogram/
plt.plot(histg)
plt.show()
```

```python
import numpy as np
```

```python
# histogram equalization
eim = cv2.equalizeHist(gim) # works only for grayscale
# why? https://stackoverflow.com/questions/15007304/histogram-equalization-not-working-on-color-image-opencv
plt.imshow(eim)
```

```python
# https://stackoverflow.com/questions/31998428/opencv-python-equalizehist-colored-image
yuvim = cv2.cvtColor(im, cv2.COLOR_BGR2YUV)
# equalize the histogram of the Y channel
yuvim[:,:,0] = cv2.equalizeHist(yuvim[:,:,0]) 
# convert the YUV image back to RGB format
img_output = cv2.cvtColor(yuvim, cv2.COLOR_YUV2BGR)
plt.imshow(img_output)
```

```python
histg = cv2.calcHist([img_output],[0],None,[256],[0,256])
plt.plot(histg)
plt.show()
```
